```python

df = pd.read_csv("imdb_top_1000.csv")
df = df.drop(columns=["Poster_Link", "Director", "Star1", "Star2", "Star3", "Star4", "Overview", "Gross", "Runtime", "Certificate", "Released_Year"])

display(df)
```

```python
import matplotlib.pyplot as plt

# Classificando o DataFrame pelos melhores indicadores
df_sorted = df.sort_values(by=['IMDB_Rating', 'Meta_score', 'No_of_Votes'], ascending=False)

# Selecionando os 10 primeiros registros
df_top_10 = df_sorted.head(10)

# Definindo os dados para o grÃ¡fico
series_titles = df_top_10['Series_Title']
imdb_ratings = df_top_10['IMDB_Rating']
meta_scores = df_top_10['Meta_score']
num_votes = df_top_10['No_of_Votes']

# Criando o grÃ¡fico
plt.figure(figsize=(12, 8))

# Plotando os dados
plt.plot(imdb_ratings, series_titles, label='IMDB Rating', marker='o', linestyle='-')
plt.plot(meta_scores, series_titles, label='Meta Score', marker='s', linestyle='--')
plt.plot(num_votes, series_titles, label='No of Votes', marker='^', linestyle='-.')

# Adicionando rÃ³tulos aos eixos e tÃ­tulo
plt.xlabel('Rating / Score / No of Votes')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies with the Best Indicators')
plt.legend()

# Exibindo o grÃ¡fico
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Classificando o DataFrame pelos melhores indicadores em ordem crescente
df_sorted = df.sort_values(by=['IMDB_Rating', 'Meta_score', 'No_of_Votes'], ascending=True)

# Selecionando os 10 primeiros registros
df_top_10 = df_sorted.head(10)

# Definindo os dados para o grÃ¡fico
series_titles = df_top_10['Series_Title']
imdb_ratings = df_top_10['IMDB_Rating']
meta_scores = df_top_10['Meta_score']
num_votes = df_top_10['No_of_Votes']

# Criando o grÃ¡fico
plt.figure(figsize=(10, 8))

# Plotando os dados em barras horizontais
plt.barh(series_titles, imdb_ratings, color='skyblue', label='IMDB Rating')
plt.barh(series_titles, meta_scores, color='lightgreen', label='Meta Score')
plt.barh(series_titles, num_votes, color='salmon', label='No of Votes')

# Adicionando rÃ³tulos aos eixos e tÃ­tulo
plt.xlabel('Rating / Score / No of Votes')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies with the Best Indicators (Ascending Order)')
plt.legend()

# Exibindo o grÃ¡fico
plt.grid(True)
plt.tight_layout()
plt.show()
```
